```python
import pandas as pd

data = pd.read_csv("../content/artwork_sample.csv")
```

```python
data.head()
```

```python
# goal is to change a column in some way
data.height.transform(lambda x: print(x))
```

```python
data.height.transform(lambda x: x)
```

```python
# convert heigh values from millimetre to cm
data.height.transform(lambda x: x/10)
```

```python
data.head()
```

```python
# transform is powerfull when combined with groupby
data.groupby('artist')
```

```python
# let's take a look at the transform function nunique, 
# which will count the number of unique values per row and per column grouped by artist. 

data.groupby('artist').transform('nunique')
```

```python
# get a single column
data.groupby('artist')['height']
```

```python
# per artist, the mean of the height column
data.groupby('artist')['height'].transform('mean')
```

```python
data.artist
```

```python
# add the new transformation column to the dataset
data["mean_height_by-artist"] = data.groupby('artist')['height'].transform('mean')

data.head()
```
