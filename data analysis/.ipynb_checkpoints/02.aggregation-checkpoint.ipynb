```python
import pandas as pd
```

```python
# importing the data into a variable dataset
data = pd.read_csv("content/artwork_sample.csv")
```

```python
# viewing the top 5 records
data.head()
```

```python
# checking the data types of the data
data.dtypes
```

```python
# viewing single column
data.acquisitionYear
```

```python
# convert to a different type example from int to float, it does not change it in the original dataset
data.acquisitionYear.astype(float)
```

```python
# as type does not change the data type in the original dataset
data.acquisitionYear.dtype
```

```python
# to persist the new data type in the original dataset
data.acquisitionYear = data.acquisitionYear.astype(float)
```

```python
data.acquisitionYear.dtype
```

```python
# AGGREGATION
```

```python
data = pd.read_csv("content/artwork_sample.csv")
```

```python
data.head()
```

```python
data.year
```

```python
# NaN are excluded in aggregation
# Aggregation on numeric data

data.year.min()
```

```python
data.year.max()
```

```python
data.year.sum()
```

```python
data.year.mean()
```

```python
data.year.std()
```

```python
# aggregation on non-numeric functions

data.artist.min()
```

```python
data.artist.max()
```

```python
data.artist.mean()
```

```python
data.artist.sum()
```

```python
# min for all columns
data.agg('min')
```

```python
# get a better sense of the data
data.agg(['min','max'])
```

```python
data.agg(['min','max','mean','std'])
```
