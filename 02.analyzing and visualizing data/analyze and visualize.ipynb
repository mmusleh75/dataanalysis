```python
import pandas as pd
#import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data/laptops.csv", encoding='ISO-8859-1', index_col=0)
```

```python
data.head()
```

```python
# how many rows and columns
data.shape
```

```python
data.drop(['Product',
           'ScreenResolution',
           'Cpu','Memory',
           'Gpu',
           'Weight'], inplace=True, axis=1)
```

```python
data.head()
```

```python
# get unique Names
data['TypeName'].unique()
```

```python
data.TypeName.value_counts()
```

```python
data.boxplot(by = 'Company', column = ['Price_euros'],
            grid = False,
            figsize = (12, 8))

plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.swarmplot(x='TypeName', y='Price_euros', data=data)

plt.title ('Price distribution by type', fontsize = 15)

#plt.xlable('Company', fontsize = 12)
plt.ylabel('Price (in euros)', fontsize = 12)

plt.show()
```
