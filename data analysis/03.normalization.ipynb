```python
# normalization: means adjust the values in a given column and change their scale
# we'll use two methods
# 1. standardization in statistics
height = data.height
height
```

```python
norm = (height - height.mean()) / height.std()
norm
```

```python
# normalize the column between 0 and 1
minmax = (height - height.min()) / (height.max() - height.min())
minmax
```

```python
minmax.min()
```

```python
minmax.max()
```

```python
data.head()
```

```python
# add the normalized values to the dataset
data.height = minmax
data.head()
```

```python
# add new column
data["normalized_height"] = norm
data.head()
```
